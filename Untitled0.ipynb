```
# 텐서 플로우 사용
import tensorflow as tf
import pandas as pd

```

```
tensordata = pd.read_csv("./data/2020typhoon(exc 10).csv")
tensordata.head()

d=tensordata[['위도','경도','기압','시속풍속','이동속도']]
j=tensordata[['위도예측','경도예측']]
print(d.shape, j.shape)

```

```
#구조
X = tf.keras.layers.Input(shape=[5])
H = tf.keras.layers.Dense(10, activation='swish')(X)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)
Y = tf.keras.layers.Dense(2)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
#학습

model.fit(d, j, epochs=100000, verbose=0)
model.fit(d, j, epochs=10)
```

```
print(model.predict(d))
```
